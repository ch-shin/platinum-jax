```python
import jax
import jax.numpy as jnp
import matplotlib.pyplot as plt
```

```python

max_iter = 2500
num_samples = 10

for r_init in range(1, 4):
    for g_init in range(1, 4):
        for c in range(1, 4):
            plt.figure()
            for key_offset in range(num_samples): 
                
                i = r_init
                j = g_init

                Xn_list = []
                for n in range(max_iter):
                    key = jax.random.PRNGKey(key_offset*max_iter + n)
                    p = jax.random.uniform(key)
                    if p < j/(i+j):
                        j += c
                    else:
                        i += c
                    Xn = j/(i+j)
                    
                    Xn_list.append(Xn)
                plt.plot(Xn_list)
                # break
            plt.xlabel('iterations')
            plt.ylabel('fraction of greens')
            plt.ylim(0, 1)
            plt.title(f"Polya's urn r={r_init}, g={g_init}, c={c}")
            plt.show()
                
```
